### *This is EDA over Flight dataset which contains some numerical and most of them are object type . We have to make it computer understandable numerical data.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

### Converting *objects* into *numeric* data .   

*   Seperate the numeric part .
*   Convert each object into number .
*   Remove the parent column .




```python
# Converting Date_of_Journey into Date , Month , Year .
# Step1 : Separting them
df['Date'] =  df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df.info()
```

```python
#Step 2 : Converting them into int
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
#Step 3 : Dropping Date_of_Journey
df.drop('Date_of_Journey',axis = 1 ,inplace = True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
# Converting Dep_Time into Dep_Hour and Deo_Min
#Step 1 : Sepearte the obj .
df['Dep_Hour'] =df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

```python
#Step 2 : Converting them into int
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
```

```python
#Step 3 : Dropping Dep_Time
df.drop('Dep_Time',axis = 1,inplace = True)
df.info()
```

```python
# Deleting the Additional data in arrival time column .
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head(2)
```

```python
# Converting Arrival_Time into Arrival_hour and Arrival_min
#Step 1 : Sepearte the obj .
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
#Step 2 : Converting them into int
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
#Step 3 : Dropping Arrival_Time
df.drop('Arrival_Time',axis = 1 ,inplace = True)
df.info()
```

```python
#Route is alrealy in source and destination , so we dont need this.
df.drop('Route',axis= 1 , inplace = True)
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df.Total_Stops.value_counts()
```

```python
# Converting Total_Stops into int by using map function and as 1 stop is the most common so nan is filled with 1.
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0 ,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4, np.nan :1})
```

```python
df.head()
```

```python
# Checking for null values .
df[df['Total_Stops'].isnull()]
```

```python
#Step 1 : Sepearte the obj into hour and min .
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head(2)
```

```python
df['Duration_hour'].unique()
```

```python
# This data is not fitted , this has anomaly .
df[df['Duration_hour'] == '5m']
```

```python
#as this data is an anomaly , so we will discard this data row
df.drop(6474,axis = 0 , inplace = True)
df.reset_index(drop = True,inplace = True)
```

```python
df['Duration_hour'].unique()
```

```python
# Step 2 : Converting them into int
df['Duration_hour'] = df['Duration_hour'].astype(int)
df.info()
```

```python
df['Duration_min'].unique()
```

```python
#Filling the null values with 0 , then converting it into int
if df['Duration_min'].isnull().any():
    df['Duration_min'] = df['Duration_min'].fillna(0)
    df['Duration_min'] = df['Duration_min'].astype(int)
df.info()
```

```python
# Step 3 : Removing the duration column
df.drop('Duration',axis = 1 , inplace = True)
# as it already transfromed into deuration_hour and duration_min
```

```python
sns.pairplot(df) # relation of these numeric values .
```

## For **Categorical data** , we have to use One hot Encoder from sklearn .


*   Intialise the One Hot Encoder .
*   Create a Dataframe of this categorical data .
*   Concatenate the dataframe to the main df .
*   Dropping the parent columns .



```python
df['Additional_Info'].unique()
```

```python
df.Additional_Info.value_counts()
```

```python
df['Airline'].unique()
```

```python
df['Destination'].unique()
```

```python
df['Source'].unique()
```

```python
#initializing the one hot encoder
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
```

```python
one_hot_encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
# getting the feature names and making a dataframe
cat_data_df = pd.DataFrame(one_hot_encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns = one_hot_encoder.get_feature_names_out())
```

```python
cat_data_df.shape
```

```python
# concating the categorical data to the main dataframe
df = pd.concat([df,cat_data_df],axis =1)
```

```python
df.info()
```

```python
# dropping the categorical columns
df.drop(['Airline','Source','Destination'], inplace = True,axis = 1)
```

```python
df.info()
```

```python
# converting the additional info column into lowercase to avoid redundancy
df['Additional_Info'] = df['Additional_Info'].str.lower()
```

```python
df['Additional_Info'].unique()
```

```python
df['Additional_Info'].isnull().sum()
```

```python
df.Additional_Info.value_counts()
```

```python
# getting the feature names and making a dataframe using one hot encoder
additional_df = pd.DataFrame(one_hot_encoder.fit_transform(df[['Additional_Info']]).toarray(),columns = one_hot_encoder.get_feature_names_out())
```

```python
# concating the additional data to the main dataframe
df = pd.concat([df,additional_df],axis =1)
df.info()
```

```python
# dropping the additional info column
df.drop('Additional_Info',axis = 1,inplace = True)
df.info()
```
